### Genral trend analysis

The target variable is 'Total_Health_Care_Expenditure'.

Expenditures on this PUF refer to payments for health care services. More specifically, expenditures in the MEPS are defined as the sum of direct payments for care provided during the year, including out-of-pocket payments and payments by private insurance, Medicaid, Medicare, and other sources.

```python

# Plot the target variable 'Total_Health_Care_Expenditure'

plt.figure(figsize=(10, 6))
sns.histplot(df['Total_Health_Care_Expenditure'], bins=30, kde=True)
plt.title('Distribution of Total Health Care Expenditure')
plt.xlabel('Total Health Care Expenditure')
plt.ylabel('Frequency')
plt.show()
```

The Total Health Care Expenditure distribution is highly right-skewed, with most values near zero and a long tail of high expenditures. This skew makes patterns hard to discern, so we apply a log transformation to spread out smaller values and compress larger ones, aiding in clearer analysis.

```python
# Apply log transformation, adding a small constant to avoid log(0)
df['Log_Total_Health_Care_Expenditure'] = np.log1p(df['Total_Health_Care_Expenditure'])

# Plot the log-transformed distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Log_Total_Health_Care_Expenditure'], bins=20, kde=True)
plt.title('Log-Transformed Distribution of Total Health Care Expenditure')
plt.xlabel('Log of Total Health Care Expenditure')
plt.ylabel('Frequency')
plt.show()
```

The log-transformed distribution of Total Health Care Expenditure shows a symmetric, bell-shaped curve, indicating that the log transformation effectively reduced the skewness of the original data. Most expenditure values now fall within a central range, approximately between log 6 and log 10, with a peak around log 8, suggesting that a majority of healthcare costs cluster around mid-range expenditure levels. This transformation makes it easier to analyze spending patterns and interpret central tendencies without extreme values overwhelming the distribution.

Notably, there is a distinct peak at zero expenditure, representing a subset of the population with no recorded healthcare spending. This group, which remains separate from the main distribution, may indicate individuals with limited access to healthcare or other unique characteristics. Accounting for these zero-cost cases separately in further analysis is essential, as they could reveal insights into healthcare access disparities or unmet healthcare needs.

For older adults, costs increase as chronic conditions progress and additional health issues arise, emphasizing the growing financial burden of chronic disease management with age. The log-transformed scale highlights this clear upward trend across age groups.

To accurately understand the relationship between family income and actual healthcare costs for chronic disease management, we excluded individuals with zero expenditure. Zero expenditure may indicate lack of access to healthcare or deferred treatment rather than an absence of healthcare needs. Including these cases could skew the analysis, as they do not reflect active healthcare spending.

```python
# Hexbin plot with log transformation for better density visualization
# Filter out negative values
df_non_negative = df[df['Family_Total_Income'] >= 0]

# Apply log transformation
df_non_negative['Log_Family_Total_Income'] = np.log1p(df_non_negative['Family_Total_Income'])

# Filter out samples with zero healthcare expenditure
df_non_zero_cost = df_non_negative[df_non_negative['Total_Health_Care_Expenditure'] > 0]

plt.figure(figsize=(10, 6))
plt.hexbin(df_non_zero_cost['Log_Family_Total_Income'], df_non_zero_cost['Log_Total_Health_Care_Expenditure'], gridsize=30, cmap="YlGnBu", mincnt=1)
plt.colorbar(label='Count')
plt.title('Log-Transformed Family Total Income vs Log-Transformed Total Health Care Expenditure')
plt.xlabel('Log of Family Total Income')
plt.ylabel('Log of Total Health Care Expenditure')
plt.show()
```

```python
# Calculate Q1 and Q3 for Family Total Income
q1 = df_non_negative['Family_Total_Income'].quantile(0.25)
q3 = df_non_negative['Family_Total_Income'].quantile(0.75)

# Create bins based on distribution
def categorize_income(income):
    if income < q1:
        return 'Low'
    elif q1 <= income <= q3:
        return 'Medium'
    else:
        return 'High'

# Apply the function to categorize income
df_non_negative['Income_Category'] = df_non_negative['Family_Total_Income'].apply(categorize_income)

# Aggregate healthcare expenditure by income category (mean or median)
category_summary = df_non_negative.groupby('Income_Category')['Total_Health_Care_Expenditure'].mean()

import seaborn as sns
import matplotlib.pyplot as plt

# Ensure the category summary is in the correct order
category_summary = category_summary.reindex(['Low', 'Medium', 'High']).reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(
    x='Income_Category',
    y='Total_Health_Care_Expenditure',
    data=category_summary,
    palette='viridis',  # Change the palette to a built-in one
    edgecolor='black'
)
plt.title('Family Total Income Category vs Average Healthcare Expenditure', fontsize=14)
plt.xlabel('Family Total Income Category', fontsize=12)
plt.ylabel('Average Healthcare Expenditure', fontsize=12)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

```

This hexbin plot shows the relationship between log-transformed family income and healthcare expenditure for chronic disease management. The highest concentration of expenditures occurs at moderate to high income levels (log income 10–12) with corresponding moderate healthcare costs (log expenditure 8–10), suggesting that families with sufficient income consistently manage healthcare costs.

Low-income families (log income near 0) are less concentrated in terms of healthcare spending, with expenditures spread out at lower levels. This may indicate potential access barriers, as these families generally show lower healthcare expenditures. Among middle- and high-income families, expenditures vary, with some incurring substantial costs, likely due to complex chronic conditions.

Overall, the plot highlights potential income-based disparities in healthcare spending, emphasizing the need for policies that support low-income families in managing chronic disease costs effectively.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure the required variables are defined
critical_diseases = [
    "Coronary_Hrt_Disease_Diagnosis", "Stroke_Diagnosis",
    "Heart_Attack_Diagnosis", "Angina_Diagnosis", "Emphysema_Diagnosis"
]

payment_types = [
    "Total_Amt_Paid_Private_Insurance", "Total_Amt_Paid_Public_Insurance",
    "Total_Amt_Paid_Out_of_pocket"
]

# Check if 'critical_comorbidity_count' column is created
df_imputed['critical_comorbidity_count'] = df_imputed[critical_diseases].sum(axis=1)

# Define income bins and critical comorbidity bins
income_bins = pd.cut(
    df_imputed["Family_Total_Income"], bins=[0, 30000, 70000, 150000], labels=["Low", "Medium", "High"]
)
df_imputed['Income_Bins'] = income_bins

critical_comorbidity_bins = pd.cut(
    df_imputed["critical_comorbidity_count"],
    bins=[0, 1, 3, df_imputed["critical_comorbidity_count"].max()],
    labels=["Low", "Medium", "High"]
)
df_imputed['Critical_Comorbidity_Bins'] = critical_comorbidity_bins

# Create Age Group bins
age_bins = pd.cut(
    df_imputed["Age"],
    bins=[50, 60, 65, 70, 100],
    labels=["50-60", "60-65", "65-70", "70+"]
)
df_imputed['Age_Group'] = age_bins

# Prepare the filtered DataFrame
filtered_df = df_imputed.copy()

# Prepare data for heatmap: Average payment grouped by Age Group, Income Bin, and Critical Comorbidity Level
heatmap_data_critical = {}
for comorbidity_level in ["Low", "Medium", "High"]:
    comorb_df = filtered_df[filtered_df["Critical_Comorbidity_Bins"] == comorbidity_level]

    # Group by Age Group and Income Bin, then calculate the mean for each payment type
    heatmap_data_critical[comorbidity_level] = comorb_df.groupby(
        ["Age_Group", "Income_Bins"]
    )[payment_types].mean()

# Set up the heatmap grid
fig, axes = plt.subplots(len(["Low", "Medium", "High"]), len(payment_types), figsize=(18, 12))

for row_idx, (comorbidity_level, data) in enumerate(heatmap_data_critical.items()):
    for col_idx, payment_type in enumerate(payment_types):
        # Pivot the data for heatmap
        pivot_data = data[payment_type].unstack()

        # Plot heatmap
        sns.heatmap(
            pivot_data,
            annot=True,
            fmt=".0f",
            cmap="YlGnBu",
            cbar=(col_idx == len(payment_types) - 1),  # Show color bar for the last column
            ax=axes[row_idx, col_idx]
        )
        axes[row_idx, col_idx].set_title(
            f"{payment_type.replace('_', ' ')}\n(Critical Comorbidity: {comorbidity_level})",
            fontsize=10
        )
        axes[row_idx, col_idx].set_xlabel("Income Bin")
        axes[row_idx, col_idx].set_ylabel("Age Group")

# Adjust layout for better readability
plt.tight_layout()
plt.show()

```

```python

filtered_df=df_imputed.copy()

age_bins = pd.cut(filtered_df["Age"], bins=[50, 60, 65, 70, 100], labels=["50-60", "60-65", "65-70", "70+"])

# Recalculate age groups and assign to filtered_df
filtered_df = filtered_df.assign(Age_Group=age_bins)

# Filter the data for high critical comorbidity
high_comorbidity_df = filtered_df[filtered_df["Critical_Comorbidity_Bins"] == "High"]

# Further filter for the specified coverage statuses
statuses_to_plot = [3, 7]  # Codes for "Uninsured (0-64)" and "Uninsured (65+)"
high_comorbidity_df = high_comorbidity_df[high_comorbidity_df["Insurance_Coverage_Status"].isin(statuses_to_plot)]

# Prepare data for heatmap: Average payment grouped by Age Group, Income Bin, and Coverage Status
heatmap_data_coverage = {}
for status_code, status_label in {3: "Uninsured (0-64)", 7: "Uninsured (65+)"}.items():
    status_df = high_comorbidity_df[high_comorbidity_df["Insurance_Coverage_Status"] == status_code]

    # Group by Age Group and Income Bin, then calculate the mean for each payment type
    heatmap_data_coverage[status_label] = status_df.groupby(["Age_Group", "Income_Bins"], observed=True)[payment_types].mean()

# Set up the heatmap grid
fig, axes = plt.subplots(len(heatmap_data_coverage), len(payment_types), figsize=(10, 8))

for row_idx, (status_label, data) in enumerate(heatmap_data_coverage.items()):
    for col_idx, payment_type in enumerate(payment_types):
        # Pivot the data for heatmap
        pivot_data = data[payment_type].unstack()

        # Check if data is not entirely NaN
        if not pivot_data.isna().all().all():
            # Plot heatmap
            sns.heatmap(
                pivot_data,
                annot=True,
                fmt=".0f",
                cmap="YlGnBu",
                cbar=(col_idx == len(payment_types) - 1),  # Show color bar for the last column
                ax=axes[row_idx, col_idx]
            )
            axes[row_idx, col_idx].set_title(f"{payment_type.replace('_', ' ')}\n(Coverage Status: {status_label})", fontsize=10)
            axes[row_idx, col_idx].set_xlabel("Income Bin")
            axes[row_idx, col_idx].set_ylabel("Age Group")
        else:
            # Remove heatmap for NaN data
            axes[row_idx, col_idx].axis("off")

# Adjust layout
plt.tight_layout()
plt.show()
```
